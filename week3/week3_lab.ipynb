```python
import heapq

class PuzzleNode:
    def __init__(self, state, parent=None, action=None, cost=0, heuristic=0):
        self.state = state
        self.parent = parent
        self.action = action
        self.cost = cost
        # Fixed typo: heruristic -> heuristic
        self.heuristic = heuristic

    # Fixed: __it__ to __lt__ (less than comparison for the priority queue)
    def __lt__(self, other):
        return (self.cost + self.heuristic) < (other.cost + other.heuristic)

# Moved helper functions outside the class for cleaner scope
def mh_dist(point1, point2):
    return abs(point1[0] - point2[0]) + abs(point1[1] - point2[1])
    
def find_pos(state, value):
    for i in range(len(state)):
        for j in range(len(state[0])):
            if state[i][j] == value:
                return (i, j)
    return None

def get_dist(init, goal):
    if len(init) != len(goal) or len(init[0]) != len(goal[0]):
        raise ValueError("States must have same dimensions")
    
    manhattan_sum = 0
    for i in range(len(init)):
        for j in range(len(init[0])):
            value = init[i][j]
            if value != 0:
                goal_pos = find_pos(goal, value)
                manhattan_sum += mh_dist((i, j), goal_pos)
    return manhattan_sum

def gen_child(node, goal_state):
    children = []
    
    # Correctly find the dynamic row and col of 0
    row = -1
    col = -1
    for i in range(len(node.state)):
        if 0 in node.state[i]:
            row = i
            col = node.state[i].index(0)
            break

    # Actions: right, down, left, up
    actions = [(0, 1), (1, 0), (0, -1), (-1, 0)]

    for action in actions:
        nrow, ncol = (row + action[0], col + action[1])

        # Check bounds
        if 0 <= nrow < 3 and 0 <= ncol < 3:
            # Deep copy the grid
            new_state = [list(r) for r in node.state]
            
            # Swap the 0 with the target tile
            new_state[row][col] = new_state[nrow][ncol]
            new_state[nrow][ncol] = 0
            
            child_node = PuzzleNode(
                new_state,
                parent=node,
                action=action,
                cost=node.cost + 1,
                heuristic=get_dist(new_state, goal_state)
            )
            children.append(child_node)
            
    return children

def a_star_search(initial_state, goal_state):
    start_node = PuzzleNode(
        initial_state, 
        heuristic=get_dist(initial_state, goal_state)
    )
    
    # Implement Priority Queue (heapq) instead of Stack
    open_set = []
    heapq.heappush(open_set, start_node)
    closed_set = set()

    while open_set:
        # heappop automatically gets the node with the lowest (cost + heuristic)
        current_node = heapq.heappop(open_set)

        if current_node.state == goal_state:
            return get_solution_path(current_node)
        
        # Add tuple representation to visited set
        closed_set.add(tuple(map(tuple, current_node.state)))

        children = gen_child(current_node, goal_state)

        for child in children:
            if tuple(map(tuple, child.state)) not in closed_set:
                heapq.heappush(open_set, child)

    return None

def get_solution_path(node):
    path = []
    while node.parent is not None:
        path.insert(0, (node.action, node.state)) # Fixed: inset -> insert
        node = node.parent # Fixed: path.parent -> node.parent
    return path

if __name__ == "__main__":
    initial_state = [[1, 2, 3], [4, 0, 5], [6, 7, 8]]
    goal_state = [[1, 2, 3], [4, 5, 6], [7, 8, 0]]
    
    solution_path = a_star_search(initial_state, goal_state)
    
    if solution_path:
        for step in solution_path:
            print(f"move {step[0]}:")
            for row in step[1]:
                print(row)
            print("-----------")
        print(f"Solution found in {len(solution_path)} moves")
    else:
        print("No solution found.")
```
